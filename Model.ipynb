# **Testing the results of the model (36 months)**

```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
import json
```

```python
x_test_36 = pd.read_csv("/content/Lending_Club/final/X_test36.csv")
y_test_36 = pd.read_csv("/content/Lending_Club/final/y_test36.csv")
```

```python
# 2. Load the coefficients from the json file to coeff_dict
with open("/content/Lending_Club/final/logistic_regression_params.json", "r") as f:
    data = json.load(f)
coeff_dict = data["coefficients_36"]

# 3. Extract the coefficients and the intercept
feature_names = [key for key in coeff_dict if key != 'intercept']
coeff_values = [coeff_dict[feat] for feat in feature_names]
intercept_36 = data['intercept_36']
intercept_value = 0.145

# # 4. Set coefficients and intercept
# modelLogR.coef_ = np.array([coeff_values])  # Shape: (1, n_features)
# modelLogR.intercept_ = np.array([intercept_value])

# # 5. Running the model
# y_pred = modelLogR.predict(x_test_36)
# y_proba = modelLogR.predict_proba(x_test_36)
```

# **Example**

```python
row = x_test_36.iloc[0]
terms = [f"({coef:.4f} * {int(val)})" for coef, val in zip(coeff_values, row)]
equation = " + ".join(terms)
intercept = intercept_value
print(f"z = {intercept:.4f} + " + equation)
z = np.dot(row, coeff_values) + intercept_value
print(f"p = 1 / (1 + exp(-z))")
p = 1 / (1 + np.exp(-z))
print(f"p = {p:.4f}")
```

# **Entire Dataset**

```python
z = np.dot(x_test_36.values, coeff_values) + intercept_36
predictions = 1 / (1 + np.exp(-z))
predictions
```

```python
x_test_36
```

```python
df = pd.DataFrame({"probabilities":predictions, "loan_status":y_test_36["loan_status"]})
```

```python
df['predictions'] = df['probabilities'].apply(lambda x: 1 if x > 0.67 else 0)
```

```python
df
```

These predictions can be used to evaulate the performance of the model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
from sklearn.metrics import confusion_matrix

# y_true: actual labels
# y_pred: predicted labels
cm = confusion_matrix(df.loan_status, df.predictions)

# Unpack confusion matrix
TN, FP, FN, TP = cm.ravel()

print(f"True Positives: {TP}")
print(f"False Positives: {FP}")
print(f"False Negatives: {FN}")
print(f"True Negatives: {TN}")
```

```python
print(classification_report(df.loan_status, df.predictions, zero_division=0))
```
