# **Testing the results of the model (36 months)**

```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
import json
```

```python
x_test_36 = pd.read_csv("path_to_your_file")
```

```python
# 1. Initializing the model
modelLogR = LogisticRegression(C=0.01,max_iter=100000, penalty='l2',tol=1e-06,
                               fit_intercept=True, class_weight="balanced", solver='lbfgs',
                               max_iter=100, random_state=666, verbose=0)

# 2. Load the coefficients from the json file to coeff_dict
with open("path_to_your_file.json", "r") as f:
    data = json.load(f)
coeff_dict = data["coefficients_36"]

# 3. Extract the coefficients and the intercept
feature_names = [key for key in coeff_dict if key != 'intercept']
coeff_values = [coeff_dict[feat] for feat in feature_names]
intercept_value = coeff_dict['intercept']

# 4. Set coefficients and intercept
modelLogR.coef_ = np.array([coeff_values])  # Shape: (1, n_features)
modelLogR.intercept_ = np.array([intercept_value])

# 5. Running the model
y_pred = modelLogR.predict(x_test_36)
y_proba = modelLogR.predict_proba(x_test_36)
```

The model performance can be evaluated using ROC AUC scores, Matthews Score and Gini Score
